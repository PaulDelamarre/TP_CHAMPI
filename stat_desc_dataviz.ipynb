```python
def texte_univariee(df, col_name, verbose=0):
    if verbose==0:
        print("La colonne \"", col_name, "\" :")
    print("De type ", type(df[col_name][0]), ".")
    print(len(np.unique(df[col_name])), "valeurs uniques (pour", df.shape[0], "lignes dans le df).")
    
    if str(type(df[col_name][0])) in ["<class 'numpy.int64'>", "<class 'numpy.float64'>"]:
        print("Indicateurs statistiques :")
        print(df[col_name].describe())
    
    
texte_univariee(df=df, col_name=df.columns[3])
```
